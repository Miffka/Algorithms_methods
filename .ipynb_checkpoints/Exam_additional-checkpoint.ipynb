```python
'''
Дополнительный экзамен по алгоритмам. "Разделяй и властвуй" и динамическое программирование.
'''

# Задача 1 - строка, только нули и единицы. Можно применять операцию инверсии к префиксу. Сколько операций нужно,
# чтобы получить строку из нулей?
# Интуитивно понятно - столько, сколько раз в префиксе линия нулей сменяется линией единиц, начиная с конца

s = input()+"0"

ans = 0

for i in range(len(s)-1):
    if s[i] != s[i+1]:
        ans += 1

print(ans)

```

```python
# Задача 2 - разбиение числа на слагаемые

n = int(input())

# n strings, k columns
k = n

P = [[0 for i in range(k+1)] for j in range(n+1)]

for i in range(n+1):
    P[i][0] = 0
for j in range(1, k+1):
    P[0][j] = 1

#print(P)

def partition(n, k):

    for i in range(1, n+1):
        for j in range(1, k+1):
            if i-j < 0:
                P[i][j] = P[i][j-1]%(10**9+7)
                continue
            P[i][j] = (P[i][j-1]+P[i-j][j])%(10**9+7)
    
    return(P[n][k])

print(partition(n,n))
```
