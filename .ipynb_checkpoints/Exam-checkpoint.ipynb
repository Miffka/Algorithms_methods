```python
# Задача 1 - по данным числам найти наименьшее натуральное число, которое делится и на a, и на b

def gcd(a, b):
    while b:
        a = a % b
        tmp = a
        a = b
        b = tmp
    return a

def lcm(a, b):
    return a*b//gcd(a,b)

int1, int2 = map(int, input().split())
print(lcm(int1, int2))
```

```python
# Задача 3 - найти минимальное число k, для которого число w можно представить как сумму чисел v1, ..., vn


# не проходит по памяти

inp = list(map(int, input().split()))
n = inp[0]
w = inp[1]
vn = inp[2:]


'''
def Coins(coinValueList,change,minCoins,coinsUsed):
    for cents in range(change+1):
        coinCount = cents
        newCoin = 1
        for j in [c for c in coinValueList if c <= cents]:
            if minCoins[cents-j] + 1 < coinCount:
                coinCount = minCoins[cents-j]+1
                newCoin = j
        minCoins[cents] = coinCount
        coinsUsed[cents] = newCoin
    return minCoins[change]

def printCoins(coinsUsed,change):
    coin = change
    while coin > 0:
        thisCoin = coinsUsed[coin]
        print(thisCoin, end = ' ')
        coin = coin - thisCoin
        
'''

'''     
def recDC(coinValueList,change,knownResults):
    # назначаем минимальным количеством монет количество монет номиналом 1
    minCoins = change
    # заводим глобальную переменную под массив разменных монеток
    global coinsUsed = []
    # если такая монета есть в наборе монет
    if change in coinValueList:
        # выводим результат 1
        knownResults[change] = 1
        # в массив разменных монет добавляем полученную монету
        coinsUsed.append(change)
        return 1
    # или если в известной таблице результатов есть какое-то число
    elif knownResults[change] > 0:
        # возвращаем его
        return knownResults[change]
    else:
        for i in [c for c in coinValueList if c <= change]:
            numCoins = 1 + recDC(coinValueList, change-i,knownResults)
            if numCoins < minCoins:
                minCoins = numCoins
                knownResults[change] = minCoins
    return minCoins
'''

def recMC(coinValueList,change,coinsUsed=[]):
    # назначаем минимальным количеством монет количество монет номиналом 1
    minCoins = change
    # если количество монет само по себе есть в массиве номиналов монет
    
    if change in coinValueList:
        #print('change in coinValueList!')
        # добавляем эту монету в массив
        coinsUsed.append(change)
        # и возвращаем количество монет - 1
        return 1
    # иначе, т.е. если такой монеты в массиве номиналов нет
    else:
        # тогда для каждого i в массиве разменных номиналов, состоящих из монет меньше суммы для размера
        for i in [c for c in coinValueList if c <= change]:
            # количество монет равно 1 плюс вызов этой функции для того же массива разменных монет
            # и сдачи номиналом текущая сумма минус номинал текущей разменной монеты
            numCoins = 1 + recMC(coinValueList,change-i,coinsUsed)
            # если полученное количество монет меньше, чем первоначальное,
            if numCoins < minCoins:
                # возвращаем новое минимальное количество монет
                minCoins = numCoins
    return minCoins

print(recMC(vn, w), end = ' ')
#print(coinsUsed)

```

```python
# Задача 6 - подсчитать количество точных вхождений второй строки в первую

str1 = input()
str2 = input()

def overlap(s1, s2):
    count = 0
    i = -1
    while True:
        i = s1.find(s2, i+1)
        if i == -1:
            return count
        count += 1

print(overlap(str1, str2))
```
