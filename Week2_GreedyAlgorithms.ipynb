```python
#Задача 1 - отрезки и точки

n = int(input()) # вводим число строк с вводом
sect = [] # инициализируем список с данными

for i in range(n):
    sect.append([int(j) for j in input().split()])

sect.sort(key=lambda i: i[1])
#print(sect)
#print(sect[0][1])
ans = []
#print("Minimal value is " + str(min(sect[2])))

# чтобы не писать условие для последнего цикла, начнем со второго отрезка
count = 1
while count != n:
    for i in range(count, n):
        #print(sect[i-1])
        
        if sect[i-1][1] >= sect[i][0]: # если правая граница предыдущего отрезка больше или равна левой границе текущего
            sect[i][1] = sect[i-1][1] # заменяем в текущем отрезке правую границу на правую границу предыдущего отрезка
            count += 1
            #print("Branch 1, i = " + str(i))
            #print("sect[i] is " + str(sect[i]))
        else:  # иначе
            ans.append(max(sect[i-1])) # добавляем в ответ какую-нибудь точку (максимум!) из пересечения текущего и предыдущего отрезка
            count += 1
            #print("Branch 2, i = " + str(i))
            
        
        if i == n-1: # если текущий отрезок - последний, то после предыдущих операций добавляем его максимум в список ответов
            ans.append(max(sect[i]))
            #print("Branch 3, i = " + str(i))
        
        #print(sect[i-1:])
        
print(len(ans))
for i in ans:
    print(i, end = " ")
```

```python
#Задача 2 - непрерывный рюкзак
n, W = input().split()
n = int(n)
W = int(W)

things = []
for i in range(n):
    things.append([int(j) for j in input().split()])
    things[i].append(things[i][0]/things[i][1])

things.sort(key = lambda i: i[2], reverse = True) # Сортируем список по третьему столбцу

# Переходим к заполнению рюкзака
priceTot = 0
for i in things:
    if i[1] <= W:
        priceTot += i[0]
        W -= i[1]
    else:
        priceTot += W/i[1]*i[0]
        W = 0
        break

print(round(priceTot, 3))
```

```python
#Задача 3 - различные слагаемые
#По данному числу n найти максимальное число k, для которого n можно представить как сумму k слагаемых
n = int(input())

"""
Сформулируем наблюдения: 
    
Сформулируем надежный шаг:
    В оптимальное решение всегда входит 1.
    В оптимальное решение для n >= 3 всегда входит 2.
    В оптимальное решение для n >= 6 всегда входит 3.
    В оптимальное решение для n >= 10 всегда входит 4.
    В оптимальное решение для n >= i-го члена прогрессии вида
        k[i] = k[i-1]+i: k(1)=1, k(2)=3, k(3)=6, ...
        эта прогрессия - ничто иное, как ряд
        k[i] = sum(range(1, i+1))
        всегда входит i-1 младших членов этой прогрессии и остаток n-k[i-1]
"""
ans = [1] #список для ответов
k = {1:1} #словарь для последовательности k
#индекс для добавления новых членов в последовательность k реализуется в цикле for
if n > 2:
    for i in range(2, n):
        k[i] = k[i-1]+i
        ans.append(i)
        if k[i] > n:
            ans.remove(i)
            ans.remove(i-1)
            ans.append(n-k[i-2])
            break

#print(k)
else:
    ans = [n]
print(len(ans))
for i in ans:
    print(i, end = " ")
```

```python
# Задача 4 - кодирование Хаффмана

s = input()
setS = set(s)
sortS = []

for i in setS:
    sortS.append([s.count(i), i])
#sortS.sort(key = lambda i: i[1], reverse = True)

#print(sortS.pop(sortS.index(min(sortS))))

vocS = {}

if len(sortS) == 1:
    vocS[sortS[0][1]] = "0"

# в этом алгоритме ошибка - необходимо брать два минимума, а не полученного родителя и минимум! - исправлено
# в коде ошибка - список удлиняется при итерировании! - непонятно, что с этим делать - исправлено при помощи флагов
while len(sortS) > 1:
    i = sortS.pop(sortS.index(min(sortS)))
    #print(i)
    j = sortS.pop(sortS.index(min(sortS)))
    #print(j)
    
    flagi, flagj = 0, 0
    
    if len(vocS) != 0:
        for key in vocS.keys():
            # необходимо добавить несколько условий дополнения словаря
            if key in i[1]:
                vocS[key] = "0" + vocS[key]
                flagi += 1
            elif key in j[1]:
                vocS[key] = "1" + vocS[key]
                flagj += 1
                
        if flagi == 0:
            vocS[i[1]] = "0"
        if flagj == 0:
            vocS[j[1]] = "1"
    else:
        vocS[i[1]] = "0"
        vocS[j[1]] = "1"
    sortS.append([i[0] + j[0], i[1] + j[1]])

#print(vocS)

ans = ""
for letter in s:
    ans += vocS[letter]


print(len(setS), len(ans))

for key, value in vocS.items():
    print(key + ": " + value)

print(ans)
```

```python
# Задача 5 - декодирование Хаффмана

codes = [] # создаем список для считывания кодировок
codesD = {} # и словарь для тех же целей в формате КОД:БУКВА
s = "" # строка с ответом
code = "" 

k, l = input().split()
k = int(k)
l = int(l)

# считываем список кодировок
for i in range(k):
    codes.append([i for i in input().split(": ")])

# читаем строку с кодировкой
sc = input()

# заполняем словарь кодировок
for i in range(k):
    codesD[codes[i][1]] = codes[i][0]

for i in sc:
    code += i
    if code in codesD.keys():
        s += codesD[code]
        code = ""

print(s)
```
